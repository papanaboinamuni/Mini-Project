# Level-1 (Univariate Analysis)

```python
# Seperating the Categorical and Numerical columns 
list_1 = Data.select_dtypes(include="object")
list_2 = Data.select_dtypes(exclude= "object")
print(tabulate({"Categorical":list_1,"Numerical":list_2},headers=["Categorical","Numerical"]))
```

```python
# Define a function to get the central tendency.
def info_of_numerical():
    for col in list_2:
        print(f"The mean of the {col} is {Data[col].mean()}")
        print(f"The median of the {col} is {Data[col].median()}")
        print(f"The mode of the {col} is {Data[col].mode()[0]}")
        print(f"The standard deviation of the {col} is {Data[col].std()}")
        print(f"Number of missing values in the {col} is {Data[col].isnull().sum()}")
        print("--------------------------------------------------------------------\n")
info_of_numerical()
```
